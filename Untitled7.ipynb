```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('go.jpeg')
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
result=DeepFace.analyze(img,actions=['gender'])
```

```python
print("Gender : ",result['gender'])
```
